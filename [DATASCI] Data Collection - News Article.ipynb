# 2. News site: Get the news articles published on March 11-12. 

### Import libraries 

```python
import requests
import json 
from bs4 import BeautifulSoup
```

### Get sources from NewsAPI

```python
url = 'http://newsapi.org/v2/everything?'
api_key = 'cfb53310f6fa41019d6d2a538be96bdd'

sources = 'http://newsapi.org/v2/sources?apiKey=c2fd8f8c95834e54ac52747f0df8318a'
requests.get(sources).json()
```

### Set the request parameters

```python
# Get articles from washingtonpost.com
parameters = {
    'domains' : 'washingtonpost.com',
    'apiKey' : api_key,
    'from' : '2021-03-11',
    'to' : '2021-03-12',
    'language' : 'en',
    'page' : '4'
}

# Store the key values pairs in a variable
parameters_json = requests.get(url, params = parameters)
parameters_json = parameters_json.json()

parameters_json
```

### Store the articles key on a variable

```python
parameters_json.keys()

news_articles = parameters_json['articles']

news_articles
```

# Save the contents into a list

```python
article_results = []

for i in range(len(news_articles)):
     # Extract full article
    article = requests.get(news_articles[i]['url'])
    soup = BeautifulSoup(article.content, 'html.parser')
    article_content = soup.find_all("div", {"class": "article-body"})
    body_text = article_content[0].get_text()
    
    results_dict = {}
    results_dict['date'] = news_articles[i]['publishedAt']
    results_dict['title'] = news_articles[i]['title']
    results_dict['full_article'] = body_text
    results_dict['author'] = news_articles[i]['author']
    
    article_results.append(results_dict)

article_results
```

# Save it into a .json file

```python
import json 
# the json file where the output must be stored 
out_file = open("news_articles.json", "w", encoding='utf8') 
  
json.dump(article_results, out_file, indent = 6, ensure_ascii=False) 
  
out_file.close() 
```
